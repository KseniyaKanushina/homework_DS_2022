1. Выведите количество пустых значений по колонкам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во пустых значений)

2.Выведите количество уникальных значений по колонокам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во уникальных значений)

```python
pd.read_sql(
    """
    SELECT 'CRIM' AS colname, count(DISTINCT crim) AS unique_values FROM boston
    UNION ALL
    SELECT 'ZN' AS colname, count(DISTINCT zn) AS unique_values FROM boston
    UNION ALL
    SELECT 'INDUS' AS colname, count(DISTINCT indus) AS unique_values FROM boston
    UNION ALL
    SELECT 'CHAS' AS colname, count(DISTINCT chas) AS unique_values FROM boston
    UNION ALL
    SELECT 'NOX' AS colname, count(DISTINCT nox) AS unique_values FROM boston
    
    """,
    con,
)
```

3. Выведите колонки, у которых медиана равна минимальному значению (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX. Напишите какой вывод можно сделать по данным в этих колонках

```python
pd.read_sql(
    """
    WITH 
        crim_cond AS (SELECT 'CRIM' as colname, avg(crim_d)=min(crim) AS cond FROM
            (SELECT crim AS crim_d FROM boston ORDER BY crim
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
  
        zn_cond AS (SELECT 'ZN' as colname, avg(zn_d)=min(zn) AS cond FROM
            (SELECT zn AS zn_d FROM boston ORDER BY zn
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
 
        indus_cond AS (SELECT 'INDUS' as colname, avg(indus_d)=min(indus) AS cond FROM
            (SELECT indus AS indus_d FROM boston ORDER BY indus
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
 
        chas_cond AS (SELECT 'CHAS' as colname, avg(chas_d)=min(chas) AS cond FROM
            (SELECT chas AS chas_d FROM boston ORDER BY chas
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
           
        nox_cond AS (SELECT 'NOX' as colname, avg(nox_d)=min(nox) AS cond FROM
            (SELECT nox AS nox_d FROM boston ORDER BY nox
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston)

                           
            
    SELECT colname FROM (
        SELECT * FROM crim_cond
        UNION ALL
        SELECT * FROM zn_cond
        UNION ALL
        SELECT * FROM indus_cond
        UNION ALL
        SELECT * FROM chas_cond
        UNION ALL
        SELECT * FROM nox_cond
        
        )
    WHERE cond=1
    
    """, 
    con,
)
```

Вывод:
Колонка CHAS - бинарная (значения 1 и 0), медиана всегда будет равна либо минимальному либо максимальному значению (за исключением случая когда кол-во разных значений (1 и 0) по выборке равно друг другу, тогда медиана равна их среднему (0.5)). Признак CHAS не сбалансирован.
Колонка ZN также не сбалансированы (кол-во одного значения данной колонки по всей выборке сильно приобладает над остальными). 

4. Выведите разницу между средним количеством комнат(RM) в домах с самой дорогой стоимостью(MEDV) и 25 самыми дешевыми домами. Аналогично по 50, 100, 200, 300 самыми дешевыми домами. (кол-во домов(25,50,100,200,300), среднее кол-во комнат в них, среднее кол-во комнат в самых дорогих, разница). Напишите влияет ли кол-во комнат на стоимость и как сильно.

```python
pd.read_sql(
    """
    WITH 
        h AS 
            (SELECT 
                avg(rm) AS avg_high
            FROM boston
            GROUP BY medv
            ORDER BY medv DESC
            LIMIT 1),
        min_25 AS 
            (SELECT 
                25 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS difference
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 25), h),
        min_50 AS 
            (SELECT 
                50 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS difference
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 50), h),
        min_100 AS 
            (SELECT 
                100 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS difference
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 100), h),
        min_200 AS 
            (SELECT 
                200 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS difference 
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 200), h),
        min_300 AS 
            (SELECT 
                300 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS difference
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 300), h)
    
    SELECT * FROM min_25
    UNION ALL
    SELECT * FROM min_50
    UNION ALL
    SELECT * FROM min_100
    UNION ALL
    SELECT * FROM min_200
    UNION ALL
    SELECT * FROM min_300
            
    """,
    con,
)
```

Вывод: количество комнат влияет на стоимость. Наблюдается положительная корреляция между этими показателями.

5.Выведите ранги значений колонки LSTAT(процент населения с более низким статусом) в домах с самой дорогой стоимостью (значение LSTAT, стоимость, ранг) среди всех знаечений LSTAT. Напишите какой вывод можно сделать по этим данным.

```python
pd.read_sql(
    """
    WITH ranked AS
        (SELECT 
            lstat,
            medv,
            RANK () OVER ( 
            ORDER BY lstat ASC
            ) lstat_rank
        FROM boston)
    SELECT 
        lstat,
        medv*1000 AS avg_price,
        lstat_rank
    FROM ranked
    WHERE medv = (SELECT max(medv) FROM boston)
    """,
    con,
)
```

Вывод: в самых дорогих домах процент населения с более низким статусом - низкий.

6.Выведите среднюю стоимость домов граничащих с рекой(CHAS) и нет (граничит/не граничит, стоимость)

```python
pd.read_sql(
    """
    WITH 
        h_river AS 
            (SELECT 
                medv,
                chas
            FROM boston
            WHERE chas = 1),
        h_common AS 
            (SELECT 
                medv,
                chas 
            FROM boston
            WHERE chas = 0)
            
    SELECT 
        avg(medv)*1000 AS avg_price,
        chas
    FROM h_common
    UNION ALL
    SELECT 
        avg(medv)*1000 AS avg_price,
        chas
    FROM h_river
    """,
    con,
)
```

7.Выведите все колонки, у которых среднее значение выше, когда дом граничит с рекой (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX. Напишите какой вывод можно сделать по этим данным.

```python
pd.read_sql(
    """
    SELECT colname FROM 
    (SELECT 
        'CRIM' AS colname, 
        (SELECT avg(crim) FROM boston WHERE chas=1)>avg(crim) AS cond
    FROM boston
    UNION ALL
    SELECT 
        'ZN' AS colname, 
        (SELECT avg(zn) FROM boston WHERE chas=1)>avg(zn) AS cond
    FROM boston
    UNION ALL
    SELECT 
        'INDUS' AS colname,
        (SELECT avg(indus) FROM boston WHERE chas=1)>avg(indus) AS cond
    FROM boston
    UNION ALL
    SELECT 
        'NOX' AS colname,
        (SELECT avg(nox) FROM boston WHERE chas=1)>avg(nox) AS cond
    FROM boston)
    WHERE cond=1
    """,
    con,
)
```
